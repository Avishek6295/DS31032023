## Check the red, green and blue color  of an image

0 = Red color
1 = Green Color
2 = Blue color

```python
img_R = img.copy()
img_G = img.copy()
img_B = img.copy()
```

```python
img_R[:, :, (1,2)] = 0
img_G[:, :, (0,2)] = 0
img_B[:, :, (0,1)] = 0
```

```python
img_rgb = np.concatenate((img_R, img_G, img_B), axis = 1)
```

```python
plt.figure(figsize = (15, 15))
plt.imshow(img_rgb)
```
