## Logistic Regression

- It is a classification Algorithm. (The target variable will be object eg: Yes or No, True or False)
- There is chance of getting confused that Logistic Regression is a Regression Algorithm. But since the target is object so hence it cannot be a Regression.

## Load the standard libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Create own data

```python
data = pd.DataFrame({
    'No. of hrs studied' : [2, 4, 6, 8, 2.5, 3.5, 4.5, 5.5, 6.5],
    'Result' : [0, 1, 0, 1, 0, 0, 1, 1, 1]
})
data
```

## Result depends on no. of hrs studied hence, Result is target and No. of hrs studied is independent

```python
sns.scatterplot(x = data['No. of hrs studied'], y = data['Result'])
```

```python
sns.regplot(x = data['No. of hrs studied'], y = data['Result'])
```
