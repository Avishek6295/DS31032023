## Create your own data

```python
data = pd.DataFrame(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
data
```

## Split the data using train test split

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size = 0.3)
```

### The main idea behind KFold Cross Validation is to create multiple train sets and test sets. Then apply ML Algorithm on each train and test set to get mulitple accuracies. This will help us to select the highest accuracy on the data and also the corresponding train and test set

## How to perform KFold Cross validation on the data

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits = 5)   ## n_splits specify the total number of train and test set to be created from the data
kf
```

```python
kf.split(data)
```

```python
list(kf.split(data))
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits = 7)   ## n_splits specify the total number of train and test set to be created from the data
kf
```

```python
list(kf.split(data))
```
