## Loading the standard libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Load the data

```python
data = pd.read_excel('Titanic.xlsx')
data.head()
```

```python
data.shape
```

## Observations:

1. Survived is the target variable

## Data Cleaning and Data Preprocessing

```python
data.isnull().sum()
```

```python
cols = ['Cabin', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14']
data = data.drop(cols, axis = 1)
data.head()
```

```python
data.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = 'median')
si
```

```python
data['Age'] = si.fit_transform(data[['Age']])
data.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = 'most_frequent')
si
```

```python
data['Embarked'] = si.fit_transform(data[['Embarked']])
data.isnull().sum()
```

```python
data.head()
```

```python
## Deleting PassengerID and Ticket from the data

data = data.drop(['PassengerId', 'Ticket'], axis = 1)
data.head()
```

```python
data = data.drop('Name', axis = 1)
data.head()
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms
```

```python
data[['Age', 'Fare']] = mms.fit_transform(data[['Age', 'Fare']])
data.head()
```

## Feature Encoding

```python
dic = {'male' : 0, 'female' : 1}
data['Sex'] = data['Sex'].replace(dic)
data.head()
```

```python
data_ohe = pd.get_dummies(data['Embarked'])
data_ohe
```

```python
data = pd.concat([data, data_ohe], axis = 1)
data = data.drop('Embarked', axis = 1)
data.head()
```

## Seperate X and y

```python
X = data.drop('Survived', axis = 1)
y = data['Survived']
```

## split the data into train and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

## Apply SVM on train set

```python
from sklearn.svm import SVC
svc = SVC()
svc
```

```python
svc.fit(X_train, y_train)
```

## Perform Predictions

```python
y_pred = svc.predict(X_test)
y_pred
```

```python
X_test
```

## Evaluations

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(svc, X, y, cv = 10)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
