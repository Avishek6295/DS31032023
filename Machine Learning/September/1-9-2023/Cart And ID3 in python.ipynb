## Import all the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Load the data

```python
data = pd.read_csv('play_tennis.csv')
data.head()
```

## Day is of no use hence delete

```python
data = data.drop('day', axis = 1)
data.head()
```

```python
dic1 = {'High' : 0, 'Normal' : 1}
dic2 = {'Weak' : 0 , 'Strong' : 1}
dic3 = {'Yes' : 1, 'No' : 0}

data['humidity'] = data['humidity'].replace(dic1)
data['wind'] = data['wind'].replace(dic2)
data['play'] = data['play'].replace(dic3)

data.head()
```

```python
data_ohe = pd.get_dummies(data[['outlook', 'temp']])
data_ohe
```

```python
data = pd.concat([data, data_ohe], axis = 1)
data.head()
```

```python
data = data.drop(['outlook', 'temp'], axis = 1)
data.head()
```

## Seperate X and y

```python
X = data.drop('play', axis = 1)
y = data['play']
```

## Apply DecisionTree Classification on the data

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()     #### the default decision tree applied on the data is Gini Index
tree
```

```python
tree.fit(X, y)
```

## Apply ID3 on the data

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion = 'entropy')
tree
```
